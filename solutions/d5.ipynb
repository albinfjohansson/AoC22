```python
orders = []
crates = []
with open("..\\input\\input5.txt", 'r') as f:
    crates_input, orders_input = (f.read().split("\n\n"))
    orders = orders_input.split("\n")
    crates = crates_input.split("\n")
```

Order the crates in a dictionary

```python
crates_build = {int(x):[] for x in crates[-1].replace(" ", "")}
# Set order in reverse so last item in list is the top most box
crate_order = crates[:-1]
crate_order.reverse()

for line in crate_order:
    for i in range(1, len(line), 4):
        if line[i] != " ":
            crates_build[int((i-1)/4+1)].append(line[i])
```
